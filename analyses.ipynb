```python
# Configurer le style des graphiques pour une meilleure esthétique
sns.set_style("whitegrid")

# --- 1. Chargement du jeu de données ---
df = pd.read_csv('Churn_Modelling.csv')

print("Jeu de données chargé. Aperçu des 5 premières lignes :\n")
print(df.head())
print("\nInformations sur le jeu de données :\n")
print(df.info())
print("\nStatistiques descriptives générales (pour toutes les colonnes numériques) :\n")
print(df.describe())

# --- 2. Séparation des variables numériques et catégorielles ---
# Détecter automatiquement les types de variables
numerical_vars = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_vars = df.select_dtypes(include=['object']).columns.tolist()

# Exclure les colonnes d'identification ou non-pertinentes de l'analyse univariée si elles ne sont pas numériques
# Par exemple, 'RowNumber', 'CustomerId', 'Surname' ne sont pas utiles pour l'analyse univariée
# Ajustez cette liste si d'autres colonnes d'ID/non-analysables existent.
vars_to_exclude = ['RowNumber', 'CustomerId', 'Surname']
numerical_vars = [var for var in numerical_vars if var not in vars_to_exclude]


# La variable 'Exited' est numérique (0 ou 1) mais doit être traitée comme catégorielle pour l'analyse univariée.
# Si 'Exited' est dans numerical_vars, déplacez-la vers categorical_vars pour le traitement approprié.
if 'Exited' in numerical_vars:
    numerical_vars.remove('Exited')
    categorical_vars.append('Exited')

print(f"\nVariables numériques détectées pour l'analyse : {numerical_vars}")
print(f"Variables catégorielles détectées pour l'analyse : {categorical_vars}")

# --- 3. Analyse Univariée des Variables Numériques ---

print("\n--- Analyse Univariée des Variables Numériques ---")
print("Statistiques descriptives détaillées (Moyenne, Médiane, Écart-type, Min, Max, Quartiles) :")
for var in numerical_vars:
    print(f"\nVariable : {var}")
    print(f"  Moyenne : {df[var].mean():.2f}")
    print(f"  Médiane : {df[var].median():.2f}")
    print(f"  Écart-type : {df[var].std():.2f}")
    print(f"  Minimum : {df[var].min():.2f}")
    print(f"  Maximum : {df[var].max():.2f}")
    print(f"  Q1 (25e percentile) : {df[var].quantile(0.25):.2f}")
    print(f"  Q3 (75e percentile) : {df[var].quantile(0.75):.2f}")
    print(f"  IQR (Q3 - Q1) : {df[var].quantile(0.75) - df[var].quantile(0.25):.2f}")


# Visualisation : Histogrammes pour les distributions
print("\nGénération des histogrammes pour les variables numériques...")
plt.figure(figsize=(18, 12))
for i, var in enumerate(numerical_vars):
    plt.subplot(2, int(len(numerical_vars)/2) + (len(numerical_vars)%2 > 0), i + 1)
    sns.histplot(df[var], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution de {var}')
    plt.xlabel(var)
    plt.ylabel('Fréquence')
plt.tight_layout()
plt.show()

# Visualisation : Boîtes à moustaches pour les valeurs aberrantes
print("\nGénération des boîtes à moustaches pour les variables numériques...")
plt.figure(figsize=(18, 12))
for i, var in enumerate(numerical_vars):
    plt.subplot(2, int(len(numerical_vars)/2) + (len(numerical_vars)%2 > 0), i + 1)
    sns.boxplot(x=df[var], color='lightgreen')
    plt.title(f'Boîte à moustaches de {var}')
    plt.xlabel(var)
plt.tight_layout()
plt.show()

# --- 4. Analyse Univariée des Variables Catégorielles ---

print("\n--- Analyse Univariée des Variables Catégorielles ---")
for var in categorical_vars:
    print(f"\nVariable : {var}")
    # Compter les fréquences
    counts = df[var].value_counts()
    # Calculer les pourcentages
    percentages = df[var].value_counts(normalize=True) * 100

    print("Fréquences absolues :")
    print(counts)
    print("\nPourcentages :")
    print(percentages.round(2)) # Arrondir à 2 décimales

    # Visualisation : Diagrammes à barres pour les distributions
    print(f"\nGénération du diagramme à barres pour {var}...")
    plt.figure(figsize=(8, 6))
    sns.countplot(x=var, data=df, palette='viridis', order=counts.index)
    plt.title(f'Distribution de {var}')
    plt.xlabel(var)
    plt.ylabel('Nombre de clients')
    # Ajouter les pourcentages au-dessus des barres
    for p in plt.gca().patches:
        height = p.get_height()
        plt.gca().text(p.get_x() + p.get_width()/2.,
                       height + 0.1,
                       f'{height/len(df)*100:.1f}%',
                       ha='center', va='bottom', fontsize=9, color='black')
    plt.tight_layout()
    plt.show()

# Cas spécifique de la variable 'Exited' pour le graphique en secteurs
if 'Exited' in categorical_vars:
    print("\nGénération du graphique en secteurs pour la variable 'Exited' (Taux de désabonnement)...")
    churn_counts = df['Exited'].value_counts()
    plt.figure(figsize=(7, 7))
    plt.pie(churn_counts, labels=['Resté (0)', 'Parti (1)'], autopct='%1.1f%%', startangle=90, colors=['#4CAF50', '#FF5733'])
    plt.title('Répartition du statut des clients (Churn)')
    plt.axis('equal') # Assure que le cercle est un cercle.
    plt.show()

print("\nAnalyse univariée terminée.")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Charger le dataset
# Assurez-vous que le fichier 'Churn_Modelling.csv' se trouve dans le même répertoire
# que votre script Python ou indiquez le chemin complet vers le fichier.
try:
    df = pd.read_csv('Churn_Modelling.csv')
    print("Dataset 'Churn_Modelling.csv' chargé avec succès.")
except FileNotFoundError:
    print("Erreur : Le fichier 'Churn_Modelling.csv' n'a pas été trouvé.")
    print("Veuillez vous assurer que le fichier est dans le même répertoire ou spécifiez le chemin complet.")
    exit() # Quitte le script si le fichier n'est pas trouvé

# --- Préparation des données pour l'analyse ---

# Renommer les colonnes pour une meilleure lisibilité dans les graphiques (optionnel)
df.rename(columns={
    'Geography': 'Pays',
    'Gender': 'Genre',
    'Age': 'Age',
    'Tenure': 'Anciennete',
    'Balance': 'Solde',
    'NumOfProducts': 'NombreProduits',
    'HasCrCard': 'AUneCarteCredit',
    'IsActiveMember': 'EstMembreActif',
    'EstimatedSalary': 'SalaireEstime',
    'Exited': 'AQuitte' # Variable cible
}, inplace=True)

# Créer une colonne pour le solde nul/non nul
df['TypeSolde'] = df['Solde'].apply(lambda x: 'Solde Nul' if x == 0 else 'Solde Non Nul')

# Créer des tranches d'âge
# Définir les bornes des tranches d'âge
bins = [18, 25, 35, 45, 55, 65, 75, df['Age'].max()]
labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65-74', '75+']
df['TrancheAge'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)


# --- Fonction pour générer des graphiques de taux de désertion ---
def plot_churn_rate(dataframe, feature, title, x_label, rotation=0):
    """
    Génère un graphique à barres montrant le taux de désertion pour une caractéristique donnée.
    """
    # Calculer le taux de désertion par catégorie de la caractéristique
    churn_rate = dataframe.groupby(feature)['AQuitte'].mean().reset_index()
    churn_rate['AQuitte'] = churn_rate['AQuitte'] * 100 # Convertir en pourcentage

    plt.figure(figsize=(10, 6))
    sns.barplot(x=feature, y='AQuitte', data=churn_rate, palette='viridis')
    plt.title(title, fontsize=16)
    plt.xlabel(x_label, fontsize=12)
    plt.ylabel('Taux de désertion (%)', fontsize=12)
    plt.ylim(0, 100) # Assurer que l'axe Y va jusqu'à 100%
    plt.xticks(rotation=rotation) # Rotation des étiquettes si nécessaire

    # Ajouter les pourcentages au-dessus des barres
    for index, row in churn_rate.iterrows():
        plt.text(index, row['AQuitte'] + 1, f"{row['AQuitte']:.1f}%", color='black', ha="center")
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

# --- Génération des graphiques ---

# 1. Taux de désertion par Pays
plot_churn_rate(df, 'Pays', 'Taux de désertion par Pays', 'Pays')


# 2. Taux de désertion par Tranche d'âge
# S'assurer que les tranches d'âge sont triées pour un affichage correct
df_sorted_age = df.sort_values('TrancheAge')
plot_churn_rate(df_sorted_age, 'TrancheAge', 'Taux de désertion par Tranche d\'âge', 'Tranche d\'âge', rotation=45)

# 2.1. Taux de désertion par Gender
# S'assurer que le genre sont triées pour un affichage correct

plot_churn_rate(df, 'Genre', 'Taux de désertion par Genre', 'Genre')


# 3. Taux de désertion par Type de Solde (Nul vs Non Nul)
plot_churn_rate(df, 'TypeSolde', 'Taux de désertion par Type de Solde', 'Type de Solde')


# 4. Taux de désertion par Nombre de Produits
plot_churn_rate(df, 'NombreProduits', 'Taux de désertion par Nombre de Produits', 'Nombre de Produits')

# 4.1 Taux de désertion par Nombre de Produits
plot_churn_rate(df, 'SalaireEstime', 'Taux de désertion par Salaire Estimé', 'SalaireEstime')

# 4.2 Taux de désertion par rapport à l'Anciennete
plot_churn_rate(df, 'Anciennete', 'Taux de désertion par rapport à l\'Anciennete', 'Anciennete')

# 4.3 Taux de désertion par rapport à A Une CarteCredit
plot_churn_rate(df, 'AUneCarteCredit', 'Taux de désertion par rapport à A Une Carte de Credit', 'AUneCarteCredit')



# 5. Taux de désertion par Statut de Membre Actif
# Remplacer les valeurs numériques par des libellés clairs pour le graphique
df['EstMembreActif_Libelle'] = df['EstMembreActif'].map({0: 'Inactif', 1: 'Actif'})
plot_churn_rate(df, 'EstMembreActif_Libelle', 'Taux de désertion par Statut de Membre', 'Statut de Membre')


print("\nLes graphiques ont été générés et affichés. Fermez les fenêtres des graphiques pour continuer.")
```

```python
# Configurer le style des graphiques pour une meilleure esthétique
sns.set_style("whitegrid")

# Charger le jeu de données
df = pd.read_csv('Churn_Modelling.csv')

# --- 1. Analyse Bivariée des Variables Numériques par rapport à 'Exited' ---

print("--- Analyse Bivariée des Variables Numériques par rapport à 'Exited' ---")

# Variables numériques à analyser
numerical_vars = ['CreditScore', 'Age', 'Balance', 'EstimatedSalary']

# Créer des boîtes à moustaches pour comparer les distributions par statut de désabonnement
plt.figure(figsize=(15, 10))
for i, var in enumerate(numerical_vars):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='Exited', y=var, data=df, palette='viridis')
    plt.title(f'Distribution de {var} par statut de dépard')
    plt.xlabel('Statut de dépard (0: Resté, 1: Parti)')
    plt.ylabel(var)
plt.tight_layout()
plt.show()

# Alternativement, utiliser des histogrammes superposés ou des KDE plots pour voir la densité
plt.figure(figsize=(15, 10))
for i, var in enumerate(numerical_vars):
    plt.subplot(2, 2, i + 1)
    sns.histplot(data=df, x=var, hue='Exited', multiple='stack', bins=30, palette='magma', alpha=0.7)
    plt.title(f'Distribution de {var} par statut de dépard')
    plt.xlabel(var)
    plt.ylabel('Fréquence')
plt.tight_layout()
plt.show()

# --- 2. Analyse Bivariée des Variables Catégorielles par rapport à 'Exited' ---

print("\n--- Analyse Bivariée des Variables Catégorielles par rapport à 'Exited' ---")

# Variables catégorielles à analyser (exclure 'Exited' elle-même comme variable explicative ici)
categorical_vars_for_bivariate = ['Geography', 'Gender', 'NumOfProducts', 'HasCrCard', 'IsActiveMember']

# Créer des diagrammes à barres pour montrer le taux de désabonnement par catégorie
plt.figure(figsize=(18, 12))
for i, var in enumerate(categorical_vars_for_bivariate):
    plt.subplot(2, 3, i + 1)
    # Calculer le taux de désabonnement moyen (proportion de '1' dans 'Exited') pour chaque catégorie
    ax = sns.barplot(x=var, y='Exited', data=df, palette='cividis')
    plt.title(f'Taux de dépard par {var}')
    plt.xlabel(var)
    plt.ylabel('Taux de dépard moyen')

    # Afficher les pourcentages sur les barres
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.2%}',
                    (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='bottom',
                    xytext=(0, 5),
                    textcoords='offset points',
                    fontsize=9, color='black')
plt.tight_layout()
plt.show()
```
